```python
with open('conv.txt', 'r') as data:
    text= data.read()
    
print(text)
    
```

```python
# removing unrecognized characters

encoded_text = text.encode("ascii", "ignore")
cleaned_text= encoded_text.decode("utf-8")
print(cleaned_text)
```

```python
## Storing Punctuation...

char ='''.?!,'''
for element in cleaned_text:
    if element in char:
        cleaned_text = cleaned_text.replace(element,"")
print(cleaned_text)
```

```python
paragraph= cleaned_text.splitlines()

print(paragraph)
```

```python
par_1=[]
char=[]
for lines in paragraph:
    sentence= lines.split(':')
    
    
    if len(sentence)!=1:
        par_1.append(sentence)
        char.append(sentence[0])
```

```python
unique_characters= pd.unique(char)
a= set(unique_characters)
n= len(unique_characters)
print(a)
```

```python
# getting details for all unique speakers in the datsets

for people in unique_characters:
    conv_words=[]
    unique_words=[]
    
    for person, speech in para_1:
        if people == person:
            conv_words+=(speech.lower()).split()
        unique_words=np.unique(conv_words)
    
    
    
    with open('I:\Python Project\Text\{}.txt'.format(people), 'a') as f:
        for words in unique_words:
            f.write(words+'\n')

    
```
